## Challenge 1
#### 1.- Define a function that simulates rolling a dice 10 times. Save the information in a dataframe.
**Hint**: you can use the *choices* function from module *random* to help you with the simulation.

```python
# your code here
def roll_dice_simulation(number_of_rolls=10):
    dice_rolls = [random.choice([1, 2, 3, 4, 5, 6]) for _ in range(number_of_rolls)]

    dice_to_df = pd.DataFrame({'Roll': range(1, number_of_rolls + 1), 'Result': dice_rolls})
    return dice_to_df

dice_rolls_dataframe = roll_dice_simulation()

dice_rolls_dataframe.head(10)
```

## Challenge 2
Now, using the dice results obtained in *challenge 1*, your are going to define some functions that will help you calculate the mean of your data in two different ways, the median and the four quartiles. 

#### 1.- Define a function that computes the mean by summing all the observations and dividing by the total number of observations. You are not allowed to use any methods or functions that directly calculate the mean value. 

```python
# your code here
def calculate_mean_manual(dice_rolls_dataframe):
    """
    Calculate the mean of a list of numbers manually.
    
    Parameters:
    data (list): A list of numerical values.
    
    Returns:
    float: Mean of the input data.
    """
    total_sum = 0
    number_of_observations = len(dice_rolls_dataframe)
 
    for number in dice_rolls_dataframe:
        total_sum += number

    mean = total_sum / number_of_observations
    
    return mean

dice_results = [4,5,1,2,1,3,1,5,2,4]
mean_manual = calculate_mean_manual(dice_results)
print("Mean (calculated manually):", mean_manual)
```

#### 2.- First, calculate the frequency distribution. Then, calculate the mean using the values of the frequency distribution you've just computed. You are not allowed to use any methods or functions that directly calculate the mean value. 

```python
# your code here

def calculate_mean_from_frequency_distribution(frequency_distribution):
    """
    Calculate the mean using frequency distribution values manually.
    
    Parameters:
    frequency_distribution (dict): A dictionary where keys are unique values and values are frequencies.
    
    Returns:
    float: Mean calculated from the frequency distribution.
    """
    total_sum = 0
    total_count = 0
    

    for value, frequency in frequency_distribution.items():
        total_sum += value * frequency
        total_count += frequency
 
    mean = total_sum / total_count
    
    return mean

frequency_distribution = {1: 3, 2: 2, 3: 1, 4: 2, 5: 2}
mean_from_frequency_distribution = calculate_mean_from_frequency_distribution(frequency_distribution)
print("Mean (calculated from frequency distribution):", mean_from_frequency_distribution)

```

#### 3.- Define a function to calculate the median. You are not allowed to use any methods or functions that directly calculate the median value. 
**Hint**: you might need to define two computation cases depending on the number of observations used to calculate the median.

```python
# your code here

def calculate_median_manual(dice_rolls_dataframe):
    """
    Calculate the median of a list of numbers manually.
    
    Parameters:
    data (list): A list of numerical values.
    
    Returns:
    float: Median of the input data.
    """
   
    sorted_data = sorted(dice_rolls_dataframe)
  
    n = len(sorted_data)
    
    if n % 2 == 1:  # Odd number of observations
        median = sorted_data[n // 2]
    else:  # Even number of observations
        middle1 = sorted_data[(n // 2) - 1]
        middle2 = sorted_data[n // 2]
        median = (middle1 + middle2) / 2
    
    return median

dice_results = [4,5,1,2,1,3,1,5,2,4]
median_manual = calculate_median_manual(dice_results)
print("Median (calculated manually):", median_manual)
```

#### 4.- Define a function to calculate the four quartiles. You can use the function you defined above to compute the median but you are not allowed to use any methods or functions that directly calculate the quartiles. 

```python
def calculate_all_quartiles_manual(dice_rolls_dataframe):
    """
    Calculate the four quartiles (Q1, Q2, Q3, Q4) of a list of numbers manually.
    
    Parameters:
    data (list): A list of numerical values.
    
    Returns:
    tuple: Quartiles Q1, Q2 (median), Q3, Q4.
    """
   
    sorted_data = sorted(dice_rolls_dataframe)
    
    #the median (Q2)
    n = len(sorted_data)
    if n % 2 == 1:
        q2 = sorted_data[n // 2]
    else:
        middle1 = sorted_data[(n // 2) - 1]
        middle2 = sorted_data[n // 2]
        q2 = (middle1 + middle2) / 2
    
    # Q1 and Q3
    lower_half = sorted_data[:n // 2]
    upper_half = sorted_data[n // 2 + (n % 2):]
    
    q1 = calculate_median_manual(lower_half)
    q3 = calculate_median_manual(upper_half)
    
    # Q4
    q4 = sorted_data[-1]
    
    return q1, q2, q3, q4

dice_results = [4,5,1,2,1,3,1,5,2,4]
q1, q2, q3, q4 = calculate_all_quartiles_manual(dice_results)
print("Q1 (25th percentile):", q1)
print("Q2 (median - 50th percentile):", q2)
print("Q3 (75th percentile):", q3)
print("Q4 (100th percentile):", q4)
```
